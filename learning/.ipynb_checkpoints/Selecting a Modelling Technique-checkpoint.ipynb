```python
import format
import buildmodel
import numpy as np

%matplotlib notebook
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score

# models
from sklearn import linear_model
from sklearn import neural_network
from sklearn import svm 
```

```python
dataset = format.load('a_1') + format.load('b_1')
df = format.toDataFrame(format.reduce(dataset))
```

```python
X = df['x'].values.reshape(-1,1)
yLeft = df['left'].values.reshape(-1,1)
yRight = df['right'].values.reshape(-1,1)
```


# Selecting a Modelling Technique 

## Linear Models 

```python
plt.figure(dpi=80, figsize=(12, 6))
plt.tight_layout()

for i, y in enumerate([yLeft, yRight]):
    
    # Ordinary least squares
    ols = linear_model.LinearRegression().fit(X,y)
    tsr = linear_model.TheilSenRegressor().fit(X,y)
    ridA = linear_model.Ridge().fit(X,y)
    ridB = linear_model.Ridge().fit(X,y)
    ridC = linear_model.Ridge().fit(X,y)
    lasA = linear_model.Lasso().fit(X,y)    

    # plot 
    ax = plt.subplot(1,2,i+1)
    ax.scatter(X,y,s=0.1,c='lightgrey')
    plt.ylim([0,100])
    plt.xlim([0,320])
    
    ols, = ax.plot(X,ols.predict(X))
    tsr, = ax.plot(X,tsr.predict(X))
    ridA, = ax.plot(X,ridA.predict(X))
    ridB, = ax.plot(X,ridB.predict(X))
    ridC, = ax.plot(X,ridC.predict(X))
    lasA, = ax.plot(X,lasA.predict(X))

    ax.legend([ols,tsr],["Ordinary Least Squares","TSR"])
```

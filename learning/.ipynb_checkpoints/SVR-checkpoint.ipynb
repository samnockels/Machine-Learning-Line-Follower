```python
import format
import buildmodel
%matplotlib notebook
import matplotlib.pyplot as plt

from sklearn import linear_model
from sklearn.svm import SVR
```

## Select Dataset

```python
linear = buildmodel.BuildModel("linear_regression", "split")
dataset = linear.load("dataset_b")
```

```python
variables = format.extract(dataset, independant=('x','y','xy'), dependant=('l','r'),dependant_match_num_blocks=True)
X = variables['independant']['x']
Y = variables['independant']['y']
yLeft, yRight = (variables['dependant']['l'],variables['dependant']['r'])
```

```python
# plt.figure(figsize=(12, 6), dpi=80,edgecolor='k')
# ax = plt.subplot(1,2,1)
# plt.scatter(X,Y,c=yLeft,cmap='winter',s=1)
# ax = plt.subplot(1,2,2)
# plt.scatter(X,Y,c=yRight,cmap='winter',s=1)
```

```python
def plot_predictions(left_model,right_model):
    xs = []
    ys = []
    left_zs = []
    right_zs = []
    regression_line_left = []
    regression_line_right = []
    for x in range(0,320,1):
        regression_line_left.append(left_model.predict([[x]])[0][0])
        regression_line_right.append(right_model.predict([[x]])[0][0])
        for y in range(0,200,1):
            xs.append(x)
            ys.append(y)
            left_zs.append(left_model.predict([[x]])[0][0])
            right_zs.append(right_model.predict([[x]])[0][0])

    plt.figure(figsize=(12, 6), dpi=80,edgecolor='k')
    ax = plt.subplot(2,2,1)
    plt.scatter(xs,ys,c=left_zs,cmap='winter',s=5,marker='s', vmin=0, vmax=100)
    ax = plt.subplot(2,2,2)
    plt.scatter(xs,ys,c=right_zs,cmap='winter',s=5,marker='s',vmin=0, vmax=100)
    ax = plt.subplot(2,2,3)
    plt.scatter(X,yLeft,c='b',s=5,marker='s')
    plt.scatter(range(0,320,1),regression_line_left,c='r',s=1)
    ax = plt.subplot(2,2,4)
    plt.scatter(X,yLeft,c='b',s=5,marker='s')
    plt.scatter(range(0,320,1),regression_line_right,c='r',s=1)
```

## Linear Regression

```python
# fit model for left & right motor
reg = linear_model.LinearRegression()
left_model  = reg.fit(X,yLeft)
reg = linear_model.LinearRegression()
right_model = reg.fit(X,yRight)

# plot model predictions
plot_predictions(left_model,right_model)
```
